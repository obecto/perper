```python
from env import ConstantEnv,TestEnv
from ray.tune.registry import register_env
from ray.rllib.agents.dqn.dqn import DQNTrainer, DEFAULT_CONFIG
import ray
import asyncio

import perper.jupyter as ppr
```

```python
try:
    ray.init(num_gpus = 0, num_cpus = 4)
except:
    ray.shutdown()
    ray.init(num_gpus = 0, num_cpus = 4)
```

```python
register_env("my_env", lambda _:TestEnv(ConstantEnv()))
```

```python
trainer_config = DEFAULT_CONFIG.copy()
trainer_config['framework'] = 'torch'
trainer_config['sample_async'] = False
```

```python
trainer_config
```

```python
trainer = DQNTrainer(trainer_config, env = 'my_env')
```

```python
%%time
best_reward = 0
last_checkpoint = 0
hall_of_fame = []
for i in range(20):
    print("Training iteration {}...".format(i))
    results = trainer.train()
    this_reward = results['episode_reward_max']
    avg_reward = results['episode_reward_mean']
    if this_reward > best_reward:
        best_reward = this_reward
        #os.remove(path)
        last_checkpoint = i + 1
        hall_of_fame.append(i+1)
        print('New best reward') 
        print(best_reward)
    if i % 10 == 0:
        print('Best Reward So Far')
        print(best_reward)
        print('Average Rewards')
        print(avg_reward)
```
