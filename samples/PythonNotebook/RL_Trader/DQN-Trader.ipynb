```python
from TradingEnv import SineMarketEnv

import numpy as np
import matplotlib.pyplot as plt
import ray
import gym
from ray.tune.registry import register_env
from ray.rllib.agents.dqn import DQNTrainer, DEFAULT_CONFIG
import pandas as pd
import gc
gc.collect()
```

```python
try:
    ray.init(num_gpus = 4, num_cpus = 4)
except:
    ray.shutdown()
    ray.init(num_gpus = 4, num_cpus = 4)
```

```python
trainer_config = DEFAULT_CONFIG.copy()
trainer_config['env_config'] = {'last_states_num': None}
trainer_config['framework'] = 'tf'
trainer_config['num_gpus'] = 0.25
trainer_config["train_batch_size"] = 12000
trainer_config['horizon'] = 1000
trainer_config['rollout_fragment_length'] = 1000
trainer_config['model']["fcnet_hiddens"] = [1024,1024]
trainer_config['model']['fcnet_activation'] = 'tanh'
trainer_config['model']['use_lstm'] = False
trainer_config['model']['lstm_cell_size'] = 256
```

```python
trainer = DQNTrainer(trainer_config, SineMarketEnv)
```

```python
best_reward = 0
hall_of_fame = []
```

```python
%%time
for i in range(1000):
    print("Training iteration {}...".format(i))
    results = trainer.train()
    this_reward = results['episode_reward_mean']
    if this_reward > best_reward:
        best_reward = this_reward
        trainer.save()
        hall_of_fame.append(i+1)
        print('This Iteration Reward')
        print(this_reward)
        print('Best Reward So Far')
        print(best_reward)
    if i % 10 == 0:
        print('This Iteration Reward')
        print(this_reward)
        print('Best Reward So Far')
        print(best_reward)
        
```
